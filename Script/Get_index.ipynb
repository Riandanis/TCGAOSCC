```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd
import collections
from itertools import combinations
from collections import Counter
from sklearn import preprocessing
from tqdm import tqdm
# import skimage
```

Get true index

```python
feature_path = '/home/DL-based-Tumor-Classification/Datasets/Newest_case/1D_sorted/'
kamus = {'OSCC':1,'Non-oral':2,'Cervical':3,'EsophagealSCC':4,'EsophagealAdenocarcinoma':5,'EsophagealCancer':6}
cases = os.listdir(feature_path)
# fold = 0
for fold in range(1,10):
    print("FOLD {}".format(fold))
    for case in tqdm(cases):
        save_path  = feature_path + case + '/img_fold{}'.format(fold)
        classes = []
        path = os.path.join(feature_path,case)
        if os.path.isdir(path) and not case.endswith(".ipynb_checkpoints") and not case.startswith("1"):
            # only target OSCC related (fold 5++)
            if case.split('+')[0]!='OSCC':
                continue
            #####
            classes_all = case.split('+')
            for kelases in classes_all:
                classes.append(kamus[kelases])
            for kelas in classes:
                most_occur_true = []
                most_occur = np.array(pd.read_csv(save_path+'/class_{}_top_800_norm.csv'.format(kelas))).squeeze()
                true_index = np.array(pd.read_csv(feature_path+'Feature_name_'+ case +'.csv')).squeeze()

                for i in most_occur:
                    if(i >= true_index.shape[0]):
                        continue
                    true_id = true_index[i][1]
                    most_occur_true.append(true_id)

                df = pd.DataFrame(most_occur_true)
                if (len(classes) == 2):
                    if not os.path.exists(feature_path+'/1Compare_reformed/sampled_fold_{}/pairwise'.format(fold)):
                        os.makedirs(feature_path+'/1Compare_reformed/sampled_fold_{}/pairwise'.format(fold))
                    df.to_csv(feature_path+'/1Compare_reformed/sampled_fold_{}/pairwise/'.format(fold)+case+'_true_index_kelas_{}.csv'.format(kelas), index=False)
                else:
                    if not os.path.exists(feature_path+'/1Compare_reformed/sampled_fold_{}/multiwise'.format(fold)):
                        os.makedirs(feature_path+'/1Compare_reformed/sampled_fold_{}/multiwise'.format(fold))
                    df.to_csv(feature_path+'/1Compare_reformed/sampled_fold_{}/multiwise/'.format(fold)+case+'_true_index_kelas_{}.csv'.format(kelas), index=False)
```

```python
feature_path = "/home/DL-based-Tumor-Classification/Datasets/Newest_case/1D/1Compare_reformed/sampled_fold_6/pairwise"
files = os.listdir(feature_path)
for target in range(1,7):
    target_file = []
    for file in files:
        path = os.path.join(feature_path,file)
        if file.endswith(".ipynb_checkpoints") or os.path.isdir(path):
                continue
        file_split = file.split("_")
        case_name = file_split[0]
        code = file_split[4]
        if str(target)+'.csv' == code:
            target_file.append(file)
    arr = list(combinations(target_file,2))
    print(len(arr))
    for element in arr:
        compare_object_1 = feature_path+'/'+ element[0]
        compare_object_2 = feature_path+'/'+ element[1]
        with open(compare_object_1, 'r') as t1, open(compare_object_2, 'r') as t2:
            fileone = t1.readlines()
            filetwo = t2.readlines()
        with open(feature_path+'/combination_2/{} - {} - {}.csv'.format(target,element[0],element[1]), 'w') as outFile:
            for line in filetwo:
                if line in fileone:
                    outFile.write(line)
```
